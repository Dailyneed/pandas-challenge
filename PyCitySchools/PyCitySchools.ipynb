```python
# Import dependencies
import pandas as pd
```

```python
# Identify the first CSV file to load
schools_info ="Resources/schools_complete.csv"

# identify the second CSV file to load
students_info = "Resources/students_complete.csv"
```

```python
# Read the first file and store into DataFrame
schools_info = pd.read_csv(schools_info)
# Read the second file and store into DataFrame
students_info = pd.read_csv(students_info)
```

```python
# Merge the two DataFrames into one using the header common to both
schools_students_info = pd.merge(students_info, schools_info, how="outer", on=["school_name"])
```

```python
schools_students_info.columns
```

```python
# Preview the new(merge) DataFrame
schools_students_info.head()
```

# District Summary

```python
# Calculate total number of unique schools
total_school = len(schools_students_info["school_name"].unique())
total_school
```

```python
# Calculate total students
total_student = len(schools_students_info["student_name"])
total_student
```

```python
# Calculate total budget
total_budget = sum(schools_students_info["budget"].unique())
total_budget
```

```python
# Calculate average math score
average_mathscore = schools_students_info["math_score"].mean()
average_mathscore
```

```python
# Calculate average reading score
average_readingscore = schools_students_info["reading_score"].mean()
average_readingscore
```

```python
# Calculate the percentage of students who passed math(Passing grade >= 70)
# % passing math
passing_mathcount = schools_students_info[(schools_students_info["math_score"] >= 70)].count()["student_name"]
passing_mathpercentage = passing_mathcount / (total_student)*100
passing_mathpercentage
```

```python
# Calculate the percentage of students who passed reading
# % passing reading
passing_readingcount = schools_students_info[(schools_students_info["reading_score"] >= 70)].count()["student_name"]
passing_readingpercentage = passing_readingcount / (total_student)*100
passing_readingpercentage
```

```python
# % overall passing (the percentage of students who passed math and reading)
pass_math_readingcount = schools_students_info[(schools_students_info["math_score"] >= 70) & (schools_students_info["reading_score"] >= 70)].count()["student_name"]
overall_passingpercentage = pass_math_readingcount / (total_student)*100
overall_passingpercentage
```

```python
# Create district summary dataframe
district_summary = pd.DataFrame({
    "Total Schools": [total_school], "Total Students": [total_student], "Total Budget": [total_budget], "Average Math Score": [average_mathscore], 
    "Average Reading Score": [average_readingscore], "% Passing Math": [passing_mathpercentage], "% Passing Reading": [passing_readingpercentage], 
    "% Overall Passing": [overall_passingpercentage]})  
```

```python
# Format district summary
district_summary["Total Students"] = district_summary["Total Students"].map('{:,}'.format)
district_summary["Total Budget"]= district_summary["Total Budget"].map('${:,.2f}'.format)                                                                       
district_summary
```

# School Summary

```python
# list school names and type
type_of_school = schools_info.set_index(["school_name"])["type"]
type_of_school.head()
```

```python
# total student count for each school
per_school_count = schools_students_info["school_name"].value_counts()
per_school_count.head()
```

```python
# Calculate total school budget
per_schoolbudget = schools_students_info.groupby(["school_name"]).mean()["budget"]
per_schoolbudget.head()
```

```python
# Calulate per student budget for schools
budget_perstudent = per_schoolbudget /per_school_count
budget_perstudent.head()
```

```python
# Calculate average math score for schools
per_school_math = schools_students_info.groupby(["school_name"]).mean()["math_score"]
per_school_math.head()
```

```python
# Calculate average reading score
per_school_reading = schools_students_info.groupby(["school_name"]).mean()["reading_score"]
per_school_reading.head()
```

```python
# Calculate the number of schools with math scores of 70 or higher
schools_passingmath = schools_students_info[schools_students_info["math_score"]>= 70]
schools_passingmath.head()
```

```python
# Calculate the number of schools with reading scores of 70 or higher
schools_passingreading = schools_students_info[schools_students_info["reading_score"]>= 70]
schools_passingreading.head()
```

```python
# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
passing_math_and_reading = schools_students_info[
    (schools_students_info["reading_score"] >= 70) & (schools_students_info["math_score"] >= 70)]
passing_math_and_reading.head()
```

```python
# % passing math (the percentage of students who passed math)
per_school_passingmath = schools_passingmath.groupby(["school_name"]).count()["student_name"] / per_school_count * 100

per_school_passingmath.head()
```

```python
# % passing reading (the percentage of students who passed reading)
per_school_passingreading = schools_passingreading.groupby(["school_name"]).count()["student_name"] / per_school_count * 100
per_school_passingreading.head()
```

```python
# % overall passing (the percentage of students who passed math AND reading)
overall_passing_rate = passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / per_school_count * 100
overall_passing_rate.head()
```

```python
per_school_summary = pd.DataFrame({ "School Type": type_of_school,
                                   "Total Students": per_school_count, "Total School Budget": per_schoolbudget,
                                   "Per Student Budget": budget_perstudent, "Average Math Score": per_school_math,
                                   "Average Reading Score": per_school_reading, "% Passing Math": per_school_passingmath,
                                   "% Passing Reading": per_school_passingreading, "% Overall Passing": overall_passing_rate
    
})
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.

# Formatting
per_school_summary["Total School Budget"]= per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"]= per_school_summary["Per Student Budget"].map("${:,.2f}".format)
per_school_summary
```

Scores by School Spending

```python
# Create a table that breaks down school performance based on average spending ranges per student
# use the code provide below to create four bins with reasonable cutoff values to group school spending.
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630- 645", "$645-680"]
                
```

```python
# Create a copy of the school summary since it has the "Per Student"
school_spending= per_school_summary.copy()
```

```python
school_spending.dtypes
```

```python
school_spending["Per Student Budget"] = school_spending["Per Student Budget"].str.replace("$", "").astype('float')
```

```python
school_spending.dtypes
```

```python
# Use 'pd.cut' to categorize spending based on the bins
pd.cut(school_spending["Per Student Budget"], bins=spending_bins, labels=labels).head()
```

```python
# Place the data series into a new column inside of the DataFrame
school_spending["Spending Ranges (Per Student)"]=pd.cut(school_spending["Per Student Budget"], bins=spending_bins, labels=labels)
```

```python
school_spending["Per Student Budget"]= school_spending["Per Student Budget"].map("${:,.2f}".format)
school_spending
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_math_scores
```

```python
spending_reading_scores = school_spending.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_reading_scores
```

```python
spending_passing_math = school_spending.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_math
```

```python
spending_passing_reading = school_spending.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
spending_passing_reading
```

```python
overall_passing_spending = school_spending.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
overall_passing_spending
```

```python
spending_summary= pd.DataFrame({"Average Math Score": spending_math_scores,
                               "Average Reading Score": spending_reading_scores,
                               "% Passing Math": spending_passing_math,
                               "% Passing Reading": spending_passing_reading,
                               "% Overall Passing": overall_passing_spending})
spending_summary
```
