```python
# Dependancies
import pandas as pd
```

```python
# Identify the first CSV file to load
schools_info ="Resources/schools_complete.csv"


```

```python
# Read the first file and store into DataFrame
schools_info = pd.read_csv(schools_info)

```

```python
# Preview the DataFrame
schools_info.head()
```

```python
# identify the second CSV file to load
students_info = "Resources/students_complete.csv"
```

```python
# Read the second file and store into DataFrame
students_info = pd.read_csv(students_info)
```

```python
# Preview the second DataFrame
students_info.head()
```

```python
# Merge the two DataFrames into one using the header common to both
schools_students_info = pd.merge(students_info, schools_info, how="left", on=["school_name", "school_name"])
schools_students_info = schools_students_info
```

```python
# Preview the new(merge) DataFrame
schools_students_info.head(30)
```

# District Summary

```python
# Calculate total number of unique schools

total_school = len(schools_students_info["school_name"].unique())
total_school
```

```python
# Calculate total students

total_student = len(schools_students_info["student_name"])
total_student
```

```python
# Calculate total budget

total_budget = sum(schools_students_info["budget"].unique())
total_budget
```

```python
# Calculate average math score

average_mathscore = schools_students_info["math_score"].mean()
average_mathscore
```

```python
# Calculate average reading score

average_readingscore = schools_students_info["reading_score"].mean()
average_readingscore
```

```python
# Calculate the percentage of students who passed math(Passing grade >= 70)
# % passing math
passing_mathcount = schools_students_info[(schools_students_info["math_score"] >= 70)].count()["student_name"]
passing_mathpercentage = passing_mathcount / (total_student)*100
passing_mathpercentage
```

```python
# Calculate the percentage of students who passed reading
# % passing reading
passing_readingcount = schools_students_info[(schools_students_info["reading_score"] >= 70)].count()["student_name"]
passing_readingpercentage = passing_readingcount / (total_student)*100
passing_readingpercentage
```

```python
# % overall passing (the percentage of students who passed math and reading)
pass_math_readingcount = schools_students_info[(schools_students_info["math_score"] >= 70) & (schools_students_info["reading_score"] >= 70)].count()["student_name"]
overall_passingpercentage = pass_math_readingcount / (total_student)*100
overall_passingpercentage
```

```python
# Create district summary dataframe
district_summary = pd.DataFrame({
    "Total Schools": [total_school], "Total Students": [total_student], "Total Budget": [total_budget], "Average Math Score": [average_mathscore], 
    "Average Reading Score": [average_readingscore], "% Passing Math": [passing_mathpercentage], "% Passing Reading": [passing_readingpercentage], 
    "% Overall Passing": [overall_passingpercentage]})
    
```

```python
# Format district summary
district_summary["Total Students"] = district_summary["Total Students"].map('{:,}'.format)
district_summary["Total Budget"]= district_summary["Total Budget"].map('${:,.2f}'.format)
                                                                       

district_summary
```

# School Summary

```python
# List school names
school_names = schools_students_info["school_name"].unique()
school_names
```

```python
# List school type
schooltype = schools_students_info["type"]
schooltype
```

```python
# list school type
#schooltype = schools_students_info.set_index(["school_name"])
#schooltype
```

```python
# total student count
per_school_count = schools_students_info["school_name"].value_counts()
per_school_count
```

```python
# Calculate total school budget
per_schoolbudget = schools_students_info.groupby(["school_name"]).mean()["budget"]
per_schoolbudget
```

```python
# Calulate per student budget
budget_perstudent = per_schoolbudget /per_school_count
budget_perstudent
```

```python
# Calculate average math score
per_school_math = schools_students_info.groupby(["school_name"]).mean()["math_score"]
per_school_math
```

```python
# Calculate average reading score
per_school_reading = schools_students_info.groupby(["school_name"]).mean()["reading_score"]
per_school_reading
```

```python

schools_passingmath = schools_students_info[schools_students_info["math_score"]>= 70]
schools_passingmath.head()
```

```python
schools_passingreading = schools_students_info[schools_students_info["reading_score"]>= 70]
schools_passingreading.head()
```

```python
passing_math_and_reading = schools_students_info[
    (schools_students_info["reading_score"] >= 70) & (schools_students_info["math_score"] >= 70)]
passing_math_and_reading
```

```python
# % passing math (the percentage of students who passed math)
per_school_passingmath = schools_passingmath.groupby(["school_name"]).count()["student_name"] / per_school_count * 100

per_school_passingmath
```

```python
# % passing reading (the percentage of students who passed reading)
per_school_passingreading = schools_passingreading.groupby(["school_name"]).count()["student_name"] / per_school_count * 100
per_school_passingreading
```

```python
# % overall passing (the percentage of students who passed math AND reading)
overall_passing_rate = passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / per_school_count * 100
overall_passing_rate
```

```python
per_school_summary = pd.DataFrame({"School Names": school_names, "School Type": schooltype,
                                   " Total Students": per_school_count, "Total School Budget": per_schoolbudget,
                                   "Per Student Budget": budget_perstudent, "Average Math Score": per_school_math,
                                   "Average Reading Score": per_school_reading, "% Passing Math": per_school_passingmath,
                                   "% Passing Reading": per_school_passingreading, " % Overall Passing": overall_passing_rate
    
})
per_school_summary
```

```python
# Format district summary
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map('{:,}'.format)
per_school_summary["Per Student Budget"]= per_school_summary["Per Student Budget"].map('${:,.2f}'.format)
                                                                    
```
